How much money does a Data Scientist make?

```python
def conditions(year):
  cond_year = df["Year"] == year
  cond_job = df["job_title"] == "Data Scientist"
  salaries = df[cond_year & cond_job]
  salaries = salaries.groupby(["Year", "job_title", "experience_level"]).agg({"salary_$": "mean"})
  return salaries
```

2020

```python
data_scientist_20 = conditions(2020)
data_scientist_20
```

2021


```python
data_scientist_21 = conditions(2021)
data_scientist_21
```

2022

```python
data_scientist_22 = conditions(2022)
data_scientist_22
```

## Report

```python
ds_salary = pd.concat([data_scientist_20, data_scientist_21, data_scientist_22])
ds_salary
```

```python
ds_salary.reset_index()
```

```python
def compare_years(p1, p2, exp, year_actual, year_old):
  if float(ds_salary.iloc[p1]) > float(ds_salary.iloc[p2]):
    percentage = round(float((ds_salary.iloc[p1] / ds_salary.iloc[p2] - 1) * 100), 2)
    print(f"In {year_actual}, Data Scientist {exp} salary increase in {percentage}% with regard to the {year_old}.")
  else:
    percentage = round(float((1 - ds_salary.iloc[p1] / ds_salary.iloc[p2]) * 100), 2)
    print(f"In {year_actual}, Data Scientist {exp} salary decreased by {percentage}% with regard to the {year_old}.")
```

```python
#2020 vs 2021
result_20_21 = compare_years(3, 0, "EN", 2021, 2020)
result_20_21
```

Face-to-face work salary vs Remote work salary

```python
def salary(year):
  cond_year = df["Year"] == year
  cond_job = df["job_title"] == "Data Scientist"
  salaries = df[cond_year & cond_job]
  salaries = salaries.groupby(["Year","job_title", "experience_level","remote_ratio"]).agg({"salary_$": "mean"})
  return salaries
```

```python
#2020
ds_20 = salary(2020)
ds_20
```

In 2020, Remote work on all 3 levels of Data Science experience has paid more money than Face-to-face work.

Conjuncture of the moment: First year of pandemic.

```python
#2021
ds_21 = salary(2021)
ds_21
```

In 2021:

*Remote work paid better for Data Scientist EN.

*Semi-face work paid better for Data Scientist MI.

*Face-to-face work paid better for Data Scientist SE.

```python
#2022
ds_22 = salary(2022)
ds_22
```

In 2022:

*Semi-face work paid better for Data Scientist EN.

*Remote work paid better for both Data Scientist MI and SE.
